# **UNEMPLOYMENT ANALYSIS WITH PYTHON**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

# **Df-dataset 1** Unemployment In India

```python
df = pd.read_csv("/content/Unemployment in India.csv")
df.head(15)
```

```python
df.duplicated().sum()
df
```

```python
df = df.drop_duplicates()
df
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.strip()
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.rename(columns={'Region': 'State'}, inplace=True)
```

```python
df = df.dropna()
```

```python
print(df.isnull().sum())
```

```python
print(df.duplicated().sum())
```

```python
average_unemployment_rate = df.groupby('State')['Estimated Unemployment Rate (%)'].mean()
```

```python
state_with_highest_unemployment = average_unemployment_rate.idxmax()
highest_unemployment_rate = average_unemployment_rate.max()
```

```python
state_with_lowest_unemployment = average_unemployment_rate.idxmin()
lowest_unemployment_rate = average_unemployment_rate.min()
```

```python
sns.set_palette("husl")
plt.figure(figsize=(10, 8))
sns.boxplot(x='Estimated Unemployment Rate (%)', y='State', data=df, palette='husl')
plt.title('Distribution of Unemployment Rate by State')
plt.xlabel('Unemployment Rate (%)')
plt.ylabel('State')
plt.show()
```

```python
sns.set_palette("husl")

plt.figure(figsize=(12, 6))
average_unemployment_rate.sort_values(ascending=False).plot(kind='bar')
plt.title("Average Unemployment Rate by State")
plt.xlabel("State")
plt.ylabel("Average Unemployment Rate (%)")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Estimated Unemployment Rate (%)'], bins=20, edgecolor='k')
plt.title("Unemployment Rate Distribution")
plt.xlabel("Estimated Unemployment Rate (%)")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.pointplot(x='State', y='Estimated Unemployment Rate (%)', data=df, ci='sd', palette='husl')
plt.title('Average Unemployment Rate by State')
plt.xlabel('State')
plt.ylabel('Average Unemployment Rate (%)')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.violinplot(x='Estimated Unemployment Rate (%)', y='State', data=df, palette='husl', inner='quartile')
plt.title('Distribution of Unemployment Rate by State')
plt.xlabel('Unemployment Rate (%)')
plt.ylabel('State')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.swarmplot(x='State', y='Estimated Unemployment Rate (%)', data=df, palette='husl')
plt.title('Unemployment Rate by State')
plt.xlabel('State')
plt.ylabel('Unemployment Rate (%)')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
colors = sns.color_palette('husl', len(average_unemployment_rate))
plt.bar(average_unemployment_rate.index, average_unemployment_rate, color=colors)
plt.title('Average Unemployment Rate by State')
plt.xlabel('State')
plt.ylabel('Average Unemployment Rate (%)')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.set_palette("tab20")

plt.figure(figsize=(12, 6))
sns.lineplot(x='Date',
             y='Estimated Unemployment Rate (%)',
             hue='State',
             data=df)
plt.title("Unemployment Rate Over Time by State")
plt.xlabel("Date")
plt.ylabel("Estimated Unemployment Rate (%)")
plt.xticks(rotation=45)
plt.legend(title='State', loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show()
```

```python
sns.set_palette("husl")

sns.pairplot(df[['Estimated Unemployment Rate (%)', 'Estimated Employed', 'Estimated Labour Participation Rate (%)']], diag_kind='kde')
plt.suptitle("Pair Plot of Numerical Columns", y=1.02)
plt.show()
```

```python
sns.set_palette("Set2")

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Estimated Unemployment Rate (%)',
                y='Estimated Labour Participation Rate (%)',
                data=df, hue='State',
                palette='Set2',
                legend='full')
plt.title('Unemployment Rate vs. Labor Participation Rate')
plt.xlabel('Estimated Unemployment Rate (%)')
plt.ylabel('Estimated Labour Participation Rate (%)')
plt.legend(title='State', loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.grid(True)
plt.show()
```

```python
fig = plt.figure(figsize=(10, 5))
sns.countplot(x='Area', data=df, palette='pastel')
plt.title('Countplot of Area')
plt.xlabel('Area')
plt.ylabel('Count')
# save the plot
plt.savefig('countplot_area.png', dpi=300)
plt.show()

fig.show()
```

```python
correlation_matrix = df[['Estimated Unemployment Rate (%)', 'Estimated Employed', 'Estimated Labour Participation Rate (%)']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
colors = sns.color_palette("Set2", n_colors=len(df['State'].unique()))
grouped_data = df.groupby(['Frequency', 'State'])['Estimated Unemployment Rate (%)'].mean().unstack()

plt.figure(figsize=(12, 6))
grouped_data.plot(kind='bar', stacked=True, color=colors)
plt.xlabel('Frequency(2019 and 2020)')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.title('Unemployment Rate by Frequency and State')
plt.legend(title='State', loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.grid(True)
plt.show()
```

```python
mean_unemployment_over_time = df.groupby('Date')['Estimated Unemployment Rate (%)'].mean()

plt.figure(figsize=(10, 6))
plt.plot(mean_unemployment_over_time.index, mean_unemployment_over_time.values, marker='o', linestyle='-')
plt.xlabel('Date')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.title('Unemployment Rate Over Time')
plt.grid(True)

plt.show()
```

```python
scatter_plot = px.scatter(df,
                         x='Estimated Employed',
                         y='Estimated Unemployment Rate (%)',
                         color='State',
                         title='Unemployment Rate vs. Estimated Employed',
                         template='plotly')
scatter_plot.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'].str.strip(), format='%d-%m-%Y')
df
```
