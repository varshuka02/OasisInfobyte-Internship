```python
!pip install scikit-learn
from sklearn.model_selection import train_test_split

train, test = train_test_split(df, test_size=0.3)
print(train.shape)
print(test.shape)
```

```python
train_X = train[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']] # taking the training data features
train_y = train.Species

test_X = test[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']] # taking test data feature
test_y = test.Species
```

```python
train_X.head()
```

```python
test_X.head()
```

```python
model = svm.SVC()
model.fit(train_X, train_y)

prediction = model.predict(test_X)
accuracy = metrics.accuracy_score(test_y, prediction) * 100

print(f'The accuracy of the SVM is: {accuracy:.2f}%')
```

```python
petal = df[['PetalLengthCm','PetalWidthCm','Species']]
sepal = df[['SepalLengthCm','SepalWidthCm','Species']]
```

```python
train_p,test_p = train_test_split(petal, test_size=0.3, random_state=0) #petals
train_x_p = train_p[['PetalWidthCm','PetalLengthCm']]
train_y_p = train_p.Species

test_x_p = test_p[['PetalWidthCm','PetalLengthCm']]
test_y_p = test_p.Species
train_s,test_s = train_test_split(sepal, test_size=0.3, random_state=0) #sepals
train_x_s = train_s[['SepalWidthCm','SepalLengthCm']]
train_y_s = train_s.Species

test_x_s = test_s[['SepalWidthCm','SepalLengthCm']]
test_y_s = test_s.Species
```

```python
model_p = svm.SVC()
model_p.fit(train_x_p, train_y_p)
prediction_p = model_p.predict(test_x_p)
accuracy_p = metrics.accuracy_score(test_y_p, prediction_p) * 100
print(f'The accuracy of the SVM using Petals is: {accuracy_p:.2f}%')


model_s = svm.SVC()
model_s.fit(train_x_s, train_y_s)
prediction_s = model_s.predict(test_x_s)
accuracy_s = metrics.accuracy_score(test_y_s, prediction_s) * 100
print(f'The accuracy of the SVM using Sepals is: {accuracy_s:.2f}%')
```
